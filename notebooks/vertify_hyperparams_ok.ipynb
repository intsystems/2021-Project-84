```python
import torch
import sys
sys.path.append('../code/')
import hyperparams
# L_val = (x-2)**2
# L_train = (y-x)**2
# solution: x = 2, y = 2
```

```python
class Net(torch.nn.Module):
    def __init__(self):
        torch.nn.Module.__init__(self)
        self.w = torch.nn.Parameter(torch.randn(1))
        
        
```

```python
def param_loss(batch, model, h):    
    return (model.w - h[0])**2

def hyperparam_loss(batch, model):
    return (model.w-2)**2


net = Net()
h = torch.nn.Parameter(torch.randn(1))
optim = torch.optim.SGD([net.w], lr=10e-2)
optim2 = torch.optim.SGD([h], lr=1e-2)
hyper_grad_calc = hyperparams.AdamHyperGradCalculator(net, param_loss,
                                                                  hyperparam_loss, optim, [h])

for _ in range(10000):
    optim2.zero_grad()
    hyper_grad_calc.calc_gradients([],[])
    optim2.step()
    optim.zero_grad()
    loss = param_loss([], net, h) 
    loss.backward()
    optim.step()
    print ('w', net.w.item())
    print ('h', h.item())
    print ('loss', loss.item())
    print ('hloss', hyperparam_loss([], net).item())
    print ('--')
```
